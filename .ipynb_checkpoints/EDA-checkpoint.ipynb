```python
def divide(x, y):
    return x / y if y else -1

def find_diff(data, columns, standard = 'problem'):
    column_list = ['feature', 'cnt', 'y_mean', 'y_median', 'y_max', 'y_0.1', 'y_0.9', 'y_std', 'n_mean', 'n_median', 'n_max', 'n_0.1', 'n_0.9', 'n_std', 
                   'diff_mean', 'diff_median', 'diff_0.1', 'diff_0.9', 'diff_std',
                   'diff_mean_ratio', 'diff_median_ratio', 'diff_0.1_ratio', 'diff_0.9_ratio', 'diff_std_Ratio', 't_test_t', 't_test_p']
    res = pd.DataFrame(columns = column_list)
    for column in tqdm(columns):
        y = data[data[standard] > 0][column]
        n = data[data[standard] == 0][column]
        cnt = len(data[data[column] > 0])
        mylist = [column, cnt, y.mean(), y.median(), y.max(), y.quantile(0.1), y.quantile(0.9), y.std(),
                          n.mean(), n.median(), n.max(), n.quantile(0.1), n.quantile(0.9), n.std(),
                          y.mean()-n.mean(), y.median()-n.median(), y.quantile(0.1)-n.quantile(0.1), 
                          y.quantile(0.9)-n.quantile(0.9), y.std() - n.std(),
                          divide(y.mean(),n.mean()), divide(y.median(),n.median()), divide(y.quantile(0.1),n.quantile(0.1)), 
                          divide(y.quantile(0.9),n.quantile(0.9)), divide(y.std(),n.std()), ttest_ind(y,n)[0], ttest_ind(y,n)[1]]
        res = res.append(pd.Series(mylist, index=column_list), ignore_index=True)
    return res
```

### 각 변수에 대한 통계분석

```python
column_list = list(train.columns)
column_list.remove('problem')
```

```python
myres = find_diff(train, column_list)
```

```python
total = myres[myres['t_test_t'].notna()].sort_values(by=['diff_mean_ratio'], ascending=False)[['feature', 'cnt', 'diff_mean_ratio', 'diff_median_ratio', 'diff_0.1_ratio', 'diff_0.9_ratio','t_test_t', 't_test_p']]
```

```python
total = total[total['t_test_p'] <= 0.05]
total = total[total['cnt'] >= 750]
```
